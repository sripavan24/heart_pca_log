```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df2=df1.copy()
num_value1=df2.select_dtypes(exclude="object").columns
```

```python
for i in num_value1:
  low=df2[i].quantile(0.20)
  high=df2[i].quantile(0.70)

  IQR=high-low

  left=low-(1.5)*IQR
  right=high+(1.5)*IQR

  df2=df2[(df2[i]>=left) & (df2[i]<=right)]
```

```python
x=df2.drop("HeartDisease",axis=1)
y=df2["HeartDisease"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=41)
```

```python
cat_value=x_train.select_dtypes(include="object").columns
num_value=x_train.select_dtypes(exclude="object").columns
```

```python
comporse1=ColumnTransformer(transformers=[
                                          ("cat_values",OneHotEncoder(handle_unknown="ignore",drop="first"),cat_value),
                                          ('num_value',StandardScaler(),num_value)
])
```

```python
pre_train=comporse1.fit_transform(x_train)
```

```python
pre_test=comporse1.transform(x_test)
```

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
```

```python
pca=PCA(n_components=0.95)
pca_train=pca.fit_transform(pre_train)
pca_test=pca.transform(pre_test)
```

```python
log=LogisticRegression()
log.fit(pca_train,y_train)
```

```python
v=log.predict(pca_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy = accuracy_score(y_test, v)
print("Accuracy:", accuracy)

```

```python

print("\nClassification Report:")
print(classification_report(y_test, v))
```

```python
sns.heatmap(confusion_matrix(y_test,v),annot=True)
```
